Задание: создайте в df новый столбец final_ride_cost, который является уменьшенным на 5% столбцом ride_cost, если rating выше 6 и максимальная скорость поездки меньше 120 км/ч. В противном случае,  ride_cost увеличивается на 5%.  В новом столбце final_ride_cost отбросьте дробную часть значения ибо это рубли!


```python
import pandas as pd

url = 'https://drive.google.com/uc?id=1eqFIZCJTkXY-PsTeS-uz_DousIH3XQu-'  
rides = pd.read_csv(url) # Открываем rides_info - информация про поездки
rides.sample(5)
```

```python
def final_ride_cost_eq(row):
  if (row['rating'] > 6) & (row['speed_max'] < 120):
    res = row['ride_cost'] - row['ride_cost'] * 0.05
  else:
    res = row['ride_cost'] + row['ride_cost'] * 0.05

  return int(res)
```

```python
rides['final_ride_cost'] = rides.apply(final_ride_cost_eq, axis=1)

rides.head()
```

# 3.1.6

Обогатите датасет rides про поездки дополнительными признаками, без использовани метода merge.

🤖 Добавьте столбец feature_1 - 75% перцентиль столбца ride_cost для каждого пользоваля (user_id).

🎃 Добавьте столбец feature_2 - (разницу максимального и минимального значения rating, деленного на среднее столбца rating для каждого пользоваля (user_id).

👽 Добавьте столбец feature_3  - количество значений ride_duration, которое больше 40 для каждой машины (car_id).

👾 Добавьте столбец feature_4 - сумму значений ride_duration, которое больше 40 для каждой машины (car_id)
Результат запишите в переменную result

```python
rides_info = pd.read_csv("https://raw.githubusercontent.com/a-milenkin/Competitive_Data_Science/main/data/rides_info.csv")
cars = pd.read_csv("https://raw.githubusercontent.com/a-milenkin/Competitive_Data_Science/main/data/car_train.csv")
```

```python
import numpy as np
```

```python
rides['ride_cost'].quantile(0.75)
```

```python
1289.034483
```

```python
# perc_func_75 = lambda x: np.percentile(x, 75)
perc_func_75 = lambda x: pd.Series.quantile(x, .75)
min_max = lambda x: (max(x) - min(x)) / (sum(x) / len(x))
# more_40 = lambda x, count: count 1 if  
def more_40(x):
  count = 0
  for el in x:
    if el > 40:
      count += 1
  return count

def sum_more_40(x):
  # s = 0
  # for el in x:
  #   if el > 40:
  #     s += 1
  # return s
  return sum([el for el in x if el > 40])

rides_info['feature_1'] = rides_info.groupby('user_id', as_index=False)['ride_cost'].transform(perc_func_75)
rides_info['feature_2'] = rides_info.groupby('user_id', as_index=False)['rating'].transform(min_max)
rides_info['feature_3'] = rides_info.groupby('car_id', as_index=False)['ride_duration'].transform(more_40)
rides_info['feature_4'] = rides_info.groupby('car_id', as_index=False)['ride_duration'].transform(sum_more_40)

result = rides_info
result
```
