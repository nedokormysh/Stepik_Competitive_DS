```python
from sklearn.metrics import f1_score
```

```python
train = pd.read_csv('https://stepik.org/media/attachments/lesson/868367/hm3_val_data.csv')
train
```

```python
train['val_target'].value_counts()
```

```python
train['val_test'] = train['val_prob'].apply(lambda x: 1 if x > 0.7 else 0)
train 
```

```python
f1_score(train['val_target'], train['val_test'])
```

```python
# np.arange(0.7, 0.8, 0.001)
```

```python
# thres = np.linspace(.7, 0.8, 50)
thres = np.arange(0.7, 0.8, 0.001)

val_f1_score = []

for el in thres:
  train['val_test'] = train['val_prob'].apply(lambda x: 1 if x > el else 0)
  f1 = f1_score(train['val_target'], train['val_test'])
  # print(f'th = {el}, f1={round(f1, 5)}')
  val_f1_score.append(f1)
```

```python
import matplotlib.pyplot as plt

y = val_f1_score
x = thres
plt.plot(x, y, 'o-')
plt.title('threshold vs. f1_score')
plt.xlabel('threshold')
plt.ylabel('validation f1_score')
```
